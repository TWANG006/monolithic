### Test the reading and displaying Zygo files.

```python
import matplotlib.pyplot as plt
import numpy as np

from monolithic.io import read_zygo_binary
from monolithic.plot import show_surface_map
from monolithic.math import remove_surface, remove_polynomials
```

```python
X, Y, Z, Xca, Yca, Zca = read_zygo_binary('../data/zygo_test.dat')
fig, ax = plt.subplots(1, 2, figsize=(9, 4))
show_surface_map(X, Y, Z, ax=ax[0], colormap='jet')
show_surface_map(Xca, Yca, Zca, ax=ax[1], colormap='jet')
plt.tight_layout()
```

```python
Zflt, _, _, = remove_surface(X, Y, Z)
Zsph, _, _, = remove_polynomials(X, Y, Z, p=2)
fig, ax = plt.subplots(1, 2, figsize=(9, 4))
show_surface_map(X, Y, Zflt, ax=ax[0], colormap='jet')
show_surface_map(X, Y, Zsph, ax=ax[1], colormap='jet')
plt.tight_layout()
print(np.nanmean(Zflt), np.nanmean(Zsph))
```
